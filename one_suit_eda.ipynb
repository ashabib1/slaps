```python
from importlib import reload
import one_suit_game
import matplotlib.pyplot as plt
import numpy as np

import one_suit_game
```

```python
wins,draws,losses = [],[],[]

for i in range(11):
    result = one_suit_game.all_games_2p(i)
    wins.append(result[0]/result[3])
    draws.append(result[1]/result[3])
    losses.append(result[2]/result[3])

plt.bar(range(len(wins)),wins,color='C2')
plt.xticks(range(len(wins)))
plt.xlabel('Number of Cards per Person')
plt.ylabel('Player 1 Winning Proportion')
plt.title('Wins for Player 1')
for i in range(len(plt.yticks()[0])-1):
    plt.axhline(y=plt.yticks()[0][i],color='black',linestyle='-',linewidth=0.1)
plt.show()

plt.bar(range(len(draws)),draws,color='C1')
plt.xticks(range(len(draws)))
plt.xlabel('Number of Cards per Person')
plt.ylabel('Player 1 Drawing Proportion')
plt.title('Draws')
for i in range(len(plt.yticks()[0])-1):
    plt.axhline(y=plt.yticks()[0][i],color='black',linestyle='-',linewidth=0.1)
plt.show()

plt.bar(range(len(losses)),losses,color='C0')
plt.xticks(range(len(losses)))
plt.xlabel('Number of Cards per Person')
plt.ylabel('Player 1 Losing Proportion')
plt.title('Losses for Player 1')
for i in range(len(plt.yticks()[0])-1):
    plt.axhline(y=plt.yticks()[0][i],color='black',linestyle='-',linewidth=0.1)
plt.show()

plt.bar(range(len(wins)),np.array(wins)+np.array(draws)+np.array(losses),label='Loss')
plt.bar(range(len(wins)),np.array(wins)+np.array(draws),label='Draw')
plt.bar(range(len(wins)),wins,label='Win')
plt.xticks(range(len(wins)))
plt.xlabel('Number of Cards per Person')
plt.ylabel('Player 1 Proportions')
plt.title('Outcomes for Player 1')
for i in range(len(plt.yticks()[0])-1):
    plt.axhline(y=plt.yticks()[0][i],color='black',linestyle='-',linewidth=0.1)
plt.legend()
plt.show()
```
